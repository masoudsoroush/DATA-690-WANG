## Section 0: Background and Goals

### 0.1. A Succinct Background

Overweight is a big health problem that can cause many other health issues both among adults and children. According to the Centers for Disease Control and Preventions (CDC), weight that is higher than what is considered as a healthy weight for a given height is described as overweight or obese. Body Mass Index, or BMI, is used as a screening tool for overweight or obesity. The BMI index is defined as follows:

Body Mass Index (BMI) is a person’s weight in kilograms divided by the square of height in meters. 

According to CDC, the range of BMI is divided into the following 4 major categories:

  1. If your BMI is less than 18.5, it falls within the **severe underweight** range.
  2. If your BMI is 18.5 to <20, it falls within the **underweight** range.
  2. If your BMI is 20 to <25, it falls within the **normal**.
  3. If your BMI is 25.0 to <30, it falls within the **overweight** range.
  4. If your BMI is 30.0 or higher, it falls within the **obese** range.

Also, according to CDC, obesity itself is divided into the following subcategories:


  * **Obese**: BMI of 30 to < 35 
  * **Severe Obese** : BMI of 35 to < 40
  * **Morbid Obese** : BMI of 40 or higher. This class is sometimes categorized as “extreme” obesity.
  
As Obesity is the root of many other diseases and health problem, it is very crucial to have a quantitative understanding of its status in U.S. and in the world. According to National Institute of Health (NIH), awareness of the importance of obesity and an urgency for action in view of the increasing numbers of people of excess weight at all ages is increasing. The focus is on the effect of obesity on conditions with high morbidity and mortality—for example, coronary heart disease, ischaemic stroke, type 2 diabetes and cancers,1 whereas the effect on osteoarthritis, back pain and other musculoskeletal problems2 is underappreciated.
  
In this project, we will study the status of obesity in U.S. and compare it with the obesity condition in other countries.

### 4.2. United States of America's BMI Status Over Time

```python
# Sorting the rows of 'US_df' by 'Year' value in the ascending manner 

US_df=US_df.sort_values(by='Year',ascending=True)
```

```python
# Bar chart of the mean BMI over time

# Data of the plot
index=np.arange(2016-1975+1)
x_list=US_df.Year.apply(str).tolist()
del x_list[::2]
year_list=tuple(x_list)
men_list=US_df[US_df['Sex']=='Men'].Mean_BMI.tolist()
women_list=US_df[US_df['Sex']=='Women'].Mean_BMI.tolist()


# Plot
fig, ax = plt.subplots(figsize=(10,12))
bar_width=0.35

rects1=ax.barh(index,men_list,bar_width,alpha=1,color='dodgerblue',
               label='Men')
rects2=ax.barh(index+bar_width,women_list,bar_width,alpha=1,color='orange',
               label='Women')

plt.title('United States Mean BMI Over Time')
plt.xlabel('Mean BMI')
plt.ylabel('Year')
plt.yticks(index+bar_width/2,year_list)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Plotting directly in pandas

# Defining an auxiliary dataframe 
x_list=US_df.Year.apply(str).tolist()
del x_list[::2]
men_list=US_df[US_df['Sex']=='Men'].Mean_BMI.tolist()
women_list=US_df[US_df['Sex']=='Women'].Mean_BMI.tolist()
d={'Year':x_list,'Men':men_list,'Women':women_list}
aux_df=pd.DataFrame(d)

# Plot
aux_df.plot.barh('Year',y=['Men','Women'],figsize=(10,12))
plt.title('United States Mean BMI Over Time')
plt.xlabel('Mean BMI')
plt.show()
```

```python
# Line plot of mean BMI for men/women over time

aux_df['Year']=aux_df.Year.apply(int)

aux_df.plot.line('Year',y=['Men','Women'],figsize=(14,6))
plt.title('United States Mean BMI Over Time')
plt.xlabel('Year')
plt.ylabel('United States Mean BMI')
plt.show()
```

**Remarks**

The above plots show that mean BMI of both men and women in the United States has increased from about 25 in 1975 to around 29 in 2016. It is worth noticing that BMI=25 is the upper limit for a healthy weight.

**Remarks**

With exception of Japan, all G7 countries have passed the upper bound for a healthy BMI as of 2016. However, the mean BMI of United States is considerably higher than the mean BMI of all other 6 coiuntries, and if the same trend continues, the mean BMI of United States will pass 30 and will enter the obese range (BMI>30). It is quite remarkable that the mean BMI for men for Japan has had a slight increase in its mean BMI over the past four decades, and is still in the healthy range (Japan's mean BMI in 2016 is less than 24!).
