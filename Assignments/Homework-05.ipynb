```python
# Importing the necessary packages.

import random
import pandas as pd
```

```python
# This function generates c distinct integers between two integers a and b.

def Generate_Rand (a,b,c):
    Randlist=[]; r=0;
    while r<c:
        Rand=random.randint(a,b)
        if Rand not in Randlist:
            Randlist.append(Rand)
            r +=1
    return sorted(Randlist)
```

```python
# The new dataframe 'Four_MD_Univ' contating exactly the above 4 institutions is constructed by 'los' command and 
# 'isin' command of pandas. 

Four_MD_Univ=New_df.loc[New_df['INSTNM'].isin(S)] 
Four_MD_Univ
```

```python
# The following represents the statistical characteristics of the 4 colleges.

Four_MD_Univ['TUITIONFEE_IN'].describe()
```
